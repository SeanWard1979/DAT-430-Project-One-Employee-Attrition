```python
# Forcing columns to int; filling errors with NaN
hr_combined['MonthlyIncome'] = pd.to_numeric(hr_combined['MonthlyIncome'], errors='coerce')
hr_combined['MonthlyRate'] = pd.to_numeric(hr_combined['MonthlyRate'], errors='coerce')
hr_combined['Education'] = pd.to_numeric(hr_combined['Education'], errors='coerce')
```

```python
# Replacing 'Male' and 'Female' values with 'M' and 'F'
def replace_values(x):
    if x == 'Male':
        return 'M'
    elif x == 'Female':
        return 'F'
    elif x == 'M':
        return 'M'
    elif x == 'F':
        return 'F'
    
hr_combined['Gender'] = hr_combined['Gender'].apply(replace_values)
hr_combined['Gender'].value_counts()
```

```python
hr_training['Attrition'].value_counts()
```

```python
hr_training['Attrition'] = hr_training['Attrition'].map({'No':0, 'Yes': 1})
hr_training['Attrition'] = hr_training['Attrition'].astype(int)
hr_training['Attrition'].value_counts()
```

```python
hr_combined['Attrition'] = hr_combined['Attrition'].map({'No':0, 'Yes': 1})
hr_combined['Attrition'] = hr_combined['Attrition'].astype(int)
hr_combined['Attrition'].value_counts()
```

```python
# Making 'Gender' a binary value (either 0 for Male or 1 for Female)
hr_training['Gender'] = hr_training['Gender'].map({'M':0, 'F':1})
hr_training['Gender'] = hr_training['Gender'].astype(int)
hr_training['Gender'].value_counts()
```

```python
# Making sure all labels are encoded as integers
label_encoder = LabelEncoder()

hr_data_all['BusinessTravel'] = label_encoder.fit_transform(hr_data_all['BusinessTravel'])
hr_data_all['Department'] = label_encoder.fit_transform(hr_data_all['Department'])
hr_data_all['EducationField'] = label_encoder.fit_transform(hr_data_all['EducationField'])
hr_data_all['Gender'] = label_encoder.fit_transform(hr_data_all['Gender'])
hr_data_all['JobRole'] = label_encoder.fit_transform(hr_data_all['JobRole'])
hr_data_all['MaritalStatus'] = label_encoder.fit_transform(hr_data_all['MaritalStatus'])
hr_data_all['OverTime'] = label_encoder.fit_transform(hr_data_all['OverTime'])
hr_data_all['Attrition'] = label_encoder.fit_transform(hr_data_all['Attrition'])
```

```python
# Set indepedent and dependent variables
x= hr_data_all.drop(columns=['Attrition'])
y= hr_data_all['Attrition']
```

```python
# Splitting dataset into training/test set
x_train, x_test, y_train, y_test = train_test_split (x, y, test_size=0.2, random_state=0)
```
